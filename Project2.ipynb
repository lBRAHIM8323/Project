```python
import numpy as np
import pandas as pd
from fuzzywuzzy import fuzz
import spacy
```

```python
amazon_ds = pd.read_csv("amazon_catalogue.csv", encoding ='latin-1')
```

```python
pd_master = pd.read_csv("ProductMaster_NEW3.csv", sep=';', error_bad_lines = False)
pd_master.shape
```

```python
pd_master.head()
```

```python
item_list = []
for items in amazon_ds['item_name']:
    items = items.upper()
    item_list.append(items)
# print(item_list)
```

```python
descs_list = []
for descs in pd_master['Product_Desc']:
    descs = str(descs)
    descs_list.append(descs)    
# print(descs_list)
```

```python
list1 = item_list
list2 = descs_list

for i in list1:
    for j in list2:
        similarity_score = fuzz.ratio(i, j)
        if similarity_score >= 90:
            print(i,"&", j, "Are similar")
```
